A race condition occurs when two or more processes are occurring simultaneously (as in parallel computation) and the order in which the operations are executed matters.  Let's say i needed to increment the value of a global variable by one with each running process.  Given 3 processes, if they are run with synchronization, each process would read the global variable, increment by 1 and write back, resulting in a final value of 3.  However, if these processes ran in parallel without sychronization, each one could potentially read an initial value of zero, increment by one and the last process to finish would ultimately write a 1 to the global variable, resulting in the wrong value.

MapReduce is a programming model and an associated implementation for processing and generating large data sets with a parallel, distributed algorithm on a cluster.

MapReduce is a part of Hadoop, as Hadoop consists of storage (HDFS) and process (MapReduce).  Hadoop uses MapReduce to distribute data and code to worker nodes (mapping) for parallel processing and combines the outputs based on unique keys (reducing)

## What is cross validation (in partiticular 10-fold cross validation)?



HW1.1  Cross validation is a model validation technique for assessing how the results of a statistical analysis will generalize to an independent dataset (i.e. solve a real problem).  With 10-fold cross validation, a dataset would be divided into 10 equal sections(folds).  One fold would be used as the validation set, and the other 9 would be used as training data.  The cross validation process is then repeated 10 times, using each fold only once as a validation set.  The 10 results from the folds can then be combined (usually averaged) to produce a single estimation.
